```python
def diveach(row):
    x = row[0]
    for x1 in row[1:]:
        if (x1 != 0):
            x = (x / x1) * 10
    return x

def divalteach(row):
    x = row[0]
    for x1 in row[::2]:
        if (x1 != 0):
            x = (x * x1)
    x2 = row[1]
    for x3 in row[1::2]:
        if (x3 != 0):
            x2 = (x2 * x3)

    return x/x2

def muleach(row):
    x = row[0]
    for x1 in row[1:]:
        if (x1 != 0):
            x = (x * x1)
    return x
def sumeach(row):
    x = row[0]
    for x1 in row[1:]:
        x = (x + x1)
    return x

def subeach(row):
    x = row[0]
    for x1 in row[1:]:
        x = (x - x1)
    return x

def subalteach(row):
    x = row[0]
    for x1 in row[::2]:
        x = (x + x1)
    x2 = row[1]
    for x3 in row[1::2]:
        x2 = (x2 + x3)        
    return x-x2


#div_df = enc_train_X_df.apply(diveach, axis=1)
sum_df = enc_train_X_df.apply(sumeach, axis=1)
mul_df = enc_train_X_df.apply(muleach, axis=1)
#sub_df = enc_train_X_df.apply(subeach, axis=1)

subalt_df = enc_train_X_df.apply(subalteach, axis=1)
divalt_df = enc_train_X_df.apply(divalteach, axis=1)

print(sum_df, mul_df, subalt_df, divalt_df)

```
